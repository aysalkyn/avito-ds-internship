```python
import umap.umap_ as umap
```

```python
D = data.drop(['label'], axis=1)
scaler = preprocessing.MinMaxScaler()
D = pd.DataFrame(scaler.fit_transform(D), columns = D.columns)
D
```

```python
n_n_values = (5, 25, 50) # Значения для n_neighbors
m_d_values = (0.1, 0.6)  # Значения для min_dist

umap_results = {} # Для хранения UMAP-признаков
umap_times = {}   # Для хранения времени выполнения


for n_neighbors_val in n_n_values:
    for min_dist_val in m_d_values:
        print(f"\nВыполнение UMAP с n_neighbors={n_neighbors_val}, min_dist={min_dist_val}...")

        umap_start_time = time.time()
        # Инициализация и обучение UMAP
        # n_components=2 для 2D-визуализации
        # random_state для воспроизводимости результатов
        reducer = umap.UMAP(n_neighbors=n_neighbors_val, min_dist=min_dist_val, random_state=123)
        UMAP_features = reducer.fit_transform(D)
        umap_end_time = time.time()

        # Сохранение результатов и времени
        umap_results[(n_neighbors_val, min_dist_val)] = UMAP_features
        umap_times[(n_neighbors_val, min_dist_val)] = umap_end_time - umap_start_time

        print(f"UMAP (n_neighbors={n_neighbors_val}, min_dist={min_dist_val}) время выполнения: {umap_times[(n_neighbors_val, min_dist_val)]:.2f} секунд")

        # --- Визуализация результатов ---
        UMAP_plot_data = pd.DataFrame(UMAP_features, columns=['x', 'y'])
        UMAP_plot_data['label'] = data['label'] # Добавляем метки для раскраски

        plt.figure(figsize=(10, 8))
        sns.scatterplot(
            x='x',
            y='y',
            hue='label',
            data=UMAP_plot_data,
            palette='bright',
            s=10, # Размер точек
            alpha=0.7 # Прозрачность для лучшей видимости перекрывающихся точек
        )
        plt.title(f'UMAP Визуализация (n_neighbors={n_neighbors_val}, min_dist={min_dist_val})', fontsize=16)
        plt.xlabel('UMAP Компонента 1', fontsize=12)
        plt.ylabel('UMAP Компонента 2', fontsize=12)
        # Размещаем легенду за пределами графика для лучшей читаемости
        plt.legend(title='Класс', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.tight_layout() # Автоматически подстраивает макет, чтобы все поместилось
        plt.show()
```

```python
print("\n--- Сводка по времени выполнения UMAP ---")
for (n_n, m_d), t in umap_times.items():
    print(f"UMAP (n_neighbors={n_n}, min_dist={m_d}): {t:.2f} секунд")
print("Как правило, UMAP значительно быстрее t-SNE, особенно на больших наборах данных. UMAP также часто лучше сохраняет общую структуру данных, в то время как t-SNE превосходно отображает кластеры, но может искажать глобальные расстояния.")
```
