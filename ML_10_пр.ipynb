```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
import time
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
data[cols_with_zero] = data[cols_with_zero].replace(0, np.nan)

df_cleaned = data.dropna()
print(f"Было строк: {len(data)}, стало: {len(df_cleaned)}")
```

```python
# Разделение на признаки и целевую переменную
X = df_cleaned.drop('Outcome', axis=1)
y = df_cleaned['Outcome']

# Разделение на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

# **Реализация бэггинга**

```python
# Создание базовой модели (решающее дерево)
base_tree = DecisionTreeClassifier(max_depth=4, random_state=42)

# Инициализация и обучение бэггинга
start_time = time.time()
bagging = BaggingClassifier(
    estimator=base_tree,
    n_estimators=50,
    max_samples=0.8,
    max_features=0.8,
    random_state=42
)
bagging.fit(X_train, y_train)
bagging_time = time.time() - start_time

# Предсказание и оценка
bagging_pred = bagging.predict(X_test)
bagging_accuracy = accuracy_score(y_test, bagging_pred)
```

```python
bagging_time
```

# **Реализация бустинга**

```python
# Инициализация и обучение бустинга (AdaBoost)
start_time = time.time()
boosting = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=3),
    n_estimators=50,
    learning_rate=0.8,
    random_state=42
)
boosting.fit(X_train, y_train)
boosting_time = time.time() - start_time

# Предсказание и оценка
boosting_pred = boosting.predict(X_test)
boosting_accuracy = accuracy_score(y_test, boosting_pred)
```

```python
boosting_time
```

# **Сравнение результатов**

```python
print("баггинг")
print(classification_report(y_test, bagging_pred))

print("бустинг")
print(classification_report(y_test, boosting_pred))
```
